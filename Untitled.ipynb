# Filter Viner

```python
import numpy as np
import matplotlib.pyplot as plt

N = 200      # Число наблюд
dNoise = 0.41   # дисперсия шума
dSignal = 5  #диспес сигнала перемещение координаты
r = 0.99     # коэффициент корреляц в модели движения от -1 до 1 это альфа 1-а = а
a = 0.9    # корелляц межд  сосед точками
en = 0.1     # дисперсия СВ в модели движения
x0=76 # начальная масса
ex = en/(1-a**2) # дисперс сигнала относит ноля 0.01 /(1- 0.81) = 0.052
#print(ex)
x = np.zeros(N) # масив нолей в кол-ве N

x[0]=0
#print(x)
# моделирование изменение веса
for i in range(1,N):#            0.9*i + normal c 0.01
    x[i]= a*x[i-1]+ np.random.normal(0,en)
    #print(x[i])
#print(x)
x += x0    # Добавляем к моделированным значениям 76 как бы неизменяющая часть измерения целая часть

# добавляем шум к наблюдениям
z = x + np.random.normal(0, dNoise, N)

# Создание матриц  a= 0.9 в степени (по модулю i - j) 01 02 03 11 12 13 и т д 
R = np.array([[a**np.abs(i-j) for j in range(N)] for i in range(N)])

# Создание деогональной матрицы где по диогонали 0.01
V = np.eye(N)* dNoise #диагональная матрица для дисперсии шума наблюдения

RVinv = np.linalg.inv(R+V/ex) # к корелляц матрицв прибавляем матрицу и вычис обратную 0.01/0.052 + матрица корелляц

mz = z. mean() # среднее шума сигнала для центрирования
#
xx =np.zeros(N) # вектор для хранения оценок

for k in range(N):
    alpha = np.dot(R [:,k], RVinv) # Умножаем построчно кореляци матрицу на обратную
    xx[k]= np.dot(alpha, (z-mz) ) + mz # умножение матрицы на центрированный шум сигнала затем снова добавляем значения

fig , ax = plt.subplots(figsize=(20,10))

ax.plot(x,color='blue',label="sig")
ax.plot(z,label='shum',color='orange')
ax.plot(xx,color='green')
#ax.set_ylabel('res')
ax.grid()
```
