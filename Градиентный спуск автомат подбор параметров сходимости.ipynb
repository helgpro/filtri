## Выбор шага сходимости автоматом

```python

#X^2-5x+5 взять производную 2Х-5

def f(x):
    return x*x-5*x+5 #Сама функция

def df(x):
    return 2*x-5      #Производная
N = 20 # Число итераций
xx = 0 #начальное значение приближение
lmd =0.3 #шаг сходимости лямда

x_plt = np.arange(0, 5.0, 0.1)# от ноля до 5 с шагом 01
f_plt = [f(x) for x in x_plt]# формируем функцию для визуализации параболы
plt.ion()#интерактив режим отображ АНИМАЦИЯ

fig, ax = plt.subplots()# Создание окна и осей
ax.grid(True)
ax.plot(x_plt, f_plt)
point = ax.scatter(xx, f(xx), c='red')

mn=100    
for x in range(N):
    #xx = xx-lmd*df(xx)
    #Выбор шага сходимости

    lmd=1/min(x+1,mn)
    xx = xx-lmd*np.sign(df(xx))
    #ani = animation.FuncAnimation(fig, xx, interval=300)
    point.set_offsets([xx,f(xx)]) # изменение аргумента на текущей итерации
    
    fig.canvas.draw()
    fig.canvas.flush_events()
    time.sleep(0.02)
ani = animation.FuncAnimation(fig, animate, frames=17, repeat=True)    
# отключение анимации и делаем синим цветом последние значение    
plt.ioff()
print(xx)
ax.scatter(xx, f(xx), c = 'blue')
plt.show()
print(xx)
```
